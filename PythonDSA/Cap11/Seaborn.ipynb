```python
import random as rdm
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import seaborn as sea
import warnings as wrn
wrn.filterwarnings('ignore')
%matplotlib inline
```

```python
dados =  sea.load_dataset('tips')
dados.head()
```

```python
sea.jointplot(data = dados, x = 'total_bill', y = 'tip', kind = 'reg')
```

```python
sea.lmplot(data = dados, x = 'total_bill', y = 'tip', col = 'smoker')
```

```python
#Valores randômicos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

#Variaveis
flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

#Dataframe
dados = pd.DataFrame({'idade': idade, 'altura':altura, 'peso': peso, 'flag_fumante': flag_fumante})

#Alimenta váriavel flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'Não Fumante'})

dados.head()
```

```python
sea.set(style = 'ticks')

sea.lmplot(x = 'altura',
           y = 'peso',
           data = dados,
           hue = 'flag_fumante',
           palette = ['tab:blue', 'tab:orange'],
           height = 7)

plt.xlabel('Altura (cm)')
plt.ylabel('Peso (Kg)')
plt.title('Relação entre Altura e Peso de Fumantes e Não Fumantes')

sea.despine()

plt.show()
```
