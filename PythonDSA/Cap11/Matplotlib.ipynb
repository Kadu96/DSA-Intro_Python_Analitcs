```python
dias = [1,2,3,4,5]
dormir = [7,8,6,77,7]
comer = [2,3,4,5,3]
trabalhar = (7,8,7,2,2)
passear = [8,5,7,8,13]

plt.stackplot(dias,dormir,comer,trabalhar,passear, colors=['m','c','r','k','b'])
plt.show()
```

```python
fatias = [7,2,2,13]
atividades = ['dormir','comer','passear','trabalhar']
cores = ['olive','lime','violet','royalblue']

plt.pie(fatias, labels = atividades, colors = cores, startangle= 90, shadow= True, explode= (0,0.2,0,0))
plt.show()
```

```python
x = linspace(0,5,10)
y = x ** 2

fig = plt.figure()

axes = fig.add_axes([0, 0, 0.8, 0.8]) # type: ignore

axes.plot(x,y,'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha')
```

```python
x = linspace(0,5,10)
y = x ** 2

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # type: ignore
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # type: ignore

axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')

axes2.plot(y,x,'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura Secundária')
```

```python
x = linspace(0,5,10)
y = x ** 2

fig, axes = plt.subplots(nrows=1, ncols=3)

for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')
    
fig.tight_layout()
```

```python
x = linspace(0,5,10)
y = x ** 2

fig, axes = plt.subplots(1,2,figsize=(10,4))

axes[0].plot(x,x**2,x,exp(x))
axes[0].set_title('Escala Padrão')
axes[0].grid(True)

axes[1].plot(x,x**2,x,exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Escala Logaritmica (y)')
axes[1].grid(color = 'b', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)
```
